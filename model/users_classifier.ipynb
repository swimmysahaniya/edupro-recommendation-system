```python
df = transactions.merge(users, on='UserID')
df = df.merge(courses, on='CourseID')
df = df.merge(teachers, on='TeacherID')

df.head()
```

```python
user_features = df.groupby('UserID').agg({
    'CourseID': 'count',
    'Amount': 'sum',
    'CourseCategory': lambda x: x.mode()[0],
    'CourseLevel': lambda x: x.mode()[0],
    'CourseRating': 'mean'
}).reset_index()

# 🔹 Diversity Score (number of unique categories explored)
diversity = df.groupby('UserID')['CourseCategory'].nunique().reset_index()
diversity.rename(columns={'CourseCategory': 'DiversityScore'}, inplace=True)

# 🔹 Enrollment Frequency (number of transactions)
frequency = df.groupby('UserID')['TransactionID'].count().reset_index()
frequency.rename(columns={'TransactionID': 'EnrollmentFrequency'}, inplace=True)

# 🔹 Learning Depth (average course level)
# First convert levels to numeric
level_map = {'Beginner': 1, 'Intermediate': 2, 'Advanced': 3}
df['CourseLevelNum'] = df['CourseLevel'].map(level_map)

learning_depth = df.groupby('UserID')['CourseLevelNum'].mean().reset_index()
learning_depth.rename(columns={'CourseLevelNum': 'LearningDepth'}, inplace=True)

# 🔗 Merge all new features
user_features = user_features.merge(diversity, on='UserID')
user_features = user_features.merge(frequency, on='UserID')
user_features = user_features.merge(learning_depth, on='UserID')

user_features.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

user_features['CourseCategory'] = le.fit_transform(user_features['CourseCategory'])
user_features['CourseLevel'] = le.fit_transform(user_features['CourseLevel'])

user_features.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X = user_features.drop('UserID', axis=1)
X_scaled = scaler.fit_transform(X)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = []

for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(range(1, 10), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

kmeans = KMeans(n_clusters=4, random_state=42)
user_features['Cluster'] = kmeans.fit_predict(X_scaled)

score = silhouette_score(X_scaled, user_features['Cluster'])
print("Silhouette Score:", score)

user_features.head()
```

```python
cluster_summary = user_features.groupby('Cluster').mean(numeric_only=True)
cluster_summary
```

```python
df['Cluster'] = df['UserID'].map(user_features.set_index('UserID')['Cluster'])
```

```python
cluster_courses = df.groupby(['Cluster', 'CourseID']).size().reset_index(name='Count')

top_courses = cluster_courses.sort_values(['Cluster', 'Count'], ascending=False)

top_courses.head(10)
```

```python
def recommend_courses(user_id):
    cluster = user_features[user_features['UserID'] == user_id]['Cluster'].values[0]
    
    recs = top_courses[top_courses['Cluster'] == cluster].head(5)
    
    recs = recs.merge(courses, on='CourseID')
    
    return recs[['CourseName', 'CourseCategory', 'CourseLevel']]
```
